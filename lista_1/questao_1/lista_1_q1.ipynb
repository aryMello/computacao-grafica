## a) Escreva a matriz com coordenadas homog√™neas do operador afim do  R¬≥ no R¬≥ da rota√ß√£o em torno de ùë† = {(ùë•, ùë¶, ùëß) ‚àà ùëÖ¬≥|ùë• = 2 ùëí ùë¶ = 1}.

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=4, suppress=True)

print("="*50)
print("OPERADORES AFINS NO R¬≥")
print("="*50)

# PROBLEMA 1: Rota√ß√£o em torno da reta s (x=2, y=1)

print("\n" + "="*50)
print("PROBLEMA 1: Rota√ß√£o em torno de s (x=2, y=1)")
print("="*50)

# Ponto na reta e dire√ß√£o (paralela ao eixo Z)
P0_s = np.array([2, 1, 0])
dir_s = np.array([0, 0, 1])
theta1 = np.pi  # 180 graus

# Matriz de rota√ß√£o em torno do eixo Z
c1 = np.cos(theta1)
s1 = np.sin(theta1)
R1 = np.array([
    [c1, -s1, 0],
    [s1,  c1, 0],
    [0,   0,  1]
])

# Vetor transla√ß√£o
t1 = P0_s - R1 @ P0_s

# Matriz homog√™nea 4x4
M1 = np.eye(4)
M1[0:3, 0:3] = R1
M1[0:3, 3] = t1

print(f"\nPonto na reta: {P0_s}")
print(f"Dire√ß√£o: {dir_s} (paralela ao eixo Z)")
print(f"√Çngulo: œÄ rad = 180¬∞")
print("\nMatriz Homog√™nea 4x4:")
print(M1)
print("\nTransforma√ß√£o: T(x) = Rx + t")
print("Matriz R:")
print(R1)
print(f"Vetor t: {t1}")
```

## b) Escreva a express√£o do operador afim do R¬≥ no R¬≥ da reflex√£o em rela√ß√£o ao  plano ùê∂ = {(ùë•, ùë¶, ùëß)| (0, 1, 0) + ùëû(‚àí 2, 1 + 3, 1 ‚àí 3) + ùëù(‚àí 1, ‚àí 1, 1)}. Escreva como uma transforma√ß√£o linear seguida de uma transla√ß√£o. 

```python
# PROBLEMA 2: Reflex√£o no plano C

print("\n" + "="*50)
print("PROBLEMA 2: Reflex√£o no plano C")
print("="*50)

# VETORES
P0_c = np.array([0, 1, 0])
v1 = np.array([-2, 4, -2])
v2 = np.array([-1, -1, 1])

print(f"\nPonto no plano P0: {P0_c}")
print(f"Vetor diretor v1: {v1}")
print(f"Vetor diretor v2: {v2}")

# Vetor normal ao plano (produto vetorial)
normal = np.cross(v1, v2)
print(f"\nNormal (v1 √ó v2): {normal}")

# Normalizar
n = normal / np.linalg.norm(normal)
print(f"Normal unit√°ria n: {n}")
print(f"Normal unit√°ria (fra√ß√µes): n = (2/‚àö14, 4/‚àö14, 6/‚àö14)")

# Matriz de reflex√£o: R = I - 2nn^T
R2 = np.eye(3) - 2 * np.outer(n, n)

# Vetor transla√ß√£o
t2 = P0_c - R2 @ P0_c

# Matriz homog√™nea 4x4
M2 = np.eye(4)
M2[0:3, 0:3] = R2
M2[0:3, 3] = t2

print("\nMatriz Homog√™nea 4x4:")
print(M2)
print("\nTransforma√ß√£o: T(x) = Rx + t")
print("Matriz R (reflex√£o):")
print(R2)
print(f"Vetor t: {t2}")
```

## c) Escreva a matriz com coordenadas homog√™neas do operador afim do R¬≥ no R¬≥ da rota√ß√£o em torno de ùê∑ = (‚àí ùë°, 1 ‚àí ùë°, ùë°), com fator de transla√ß√£o do eixo (no sentido do eixo) de de 2/œÄ 

```python
# PROBLEMA 3: Rota√ß√£o Helicoidal em torno de D

print("\n" + "="*50)
print("PROBLEMA 3: Rota√ß√£o Helicoidal em torno de D")
print("="*50)

# Reta D: (-t, 1-t, t) = (0,1,0) + t(-1,-1,1)
P0_d = np.array([0, 1, 0])
v_dir = np.array([-1, -1, 1])
v_unit = v_dir / np.linalg.norm(v_dir)

theta3 = np.pi  # 180 graus
h = 2/np.pi  # fator de transla√ß√£o

print(f"\nPonto no eixo P0: {P0_d}")
print(f"Dire√ß√£o: {v_dir}")
print(f"Dire√ß√£o unit√°ria: {v_unit}")
print(f"Dire√ß√£o unit√°ria (fra√ß√µes): (-1/‚àö3, -1/‚àö3, 1/‚àö3)")
print(f"√Çngulo: œÄ rad = 180¬∞")
print(f"Fator de transla√ß√£o: h = 2/œÄ = {h:.4f}")

# F√≥rmula de Rodrigues para rota√ß√£o em torno de eixo arbitr√°rio
c3 = np.cos(theta3)
s3 = np.sin(theta3)
C = 1 - c3
vx, vy, vz = v_unit

R3 = np.array([
    [vx*vx*C + c3,     vx*vy*C - vz*s3, vx*vz*C + vy*s3],
    [vy*vx*C + vz*s3,  vy*vy*C + c3,    vy*vz*C - vx*s3],
    [vz*vx*C - vy*s3,  vz*vy*C + vx*s3, vz*vz*C + c3]
])

# Transla√ß√£o: rota√ß√£o + movimento helicoidal ao longo do eixo
t3 = P0_d - R3 @ P0_d + h * theta3 * v_unit

# Matriz homog√™nea 4x4
M3 = np.eye(4)
M3[0:3, 0:3] = R3
M3[0:3, 3] = t3

print("\nMatriz Homog√™nea 4x4:")
print(M3)
print("\nTransforma√ß√£o: T(x) = Rx + t (com componente helicoidal)")
print("Matriz R (Rodrigues):")
print(R3)
print(f"Vetor t: {t3}")
```

## VISUALIZA√á√ÉO 2D

```python
# Proje√ß√£o XY

print("\n" + "="*50)
print("GERANDO VISUALIZA√á√ïES 2D...")
print("="*50)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# V√©rtices do cubo (igual ao JavaScript)
cubo = np.array([
    [-0.5, -0.5, -0.5], [0.5, -0.5, -0.5],
    [0.5, 0.5, -0.5], [-0.5, 0.5, -0.5],
    [-0.5, -0.5, 0.5], [0.5, -0.5, 0.5],
    [0.5, 0.5, 0.5], [-0.5, 0.5, 0.5]
])

arestas = [
    [0,1], [1,2], [2,3], [3,0],  # face traseira
    [4,5], [5,6], [6,7], [7,4],  # face frontal
    [0,4], [1,5], [2,6], [3,7]   # conex√µes
]

transformacoes = [
    ("Problema 1: Rota√ß√£o em torno de s\n(x=2, y=1)", M1, 's'),
    ("Problema 2: Reflex√£o no plano C", M2, 'plane'),
    ("Problema 3: Rota√ß√£o Helicoidal\nem torno de D", M3, 'axis')
]

for idx, (titulo, matriz, ref_type) in enumerate(transformacoes):
    ax = axes[idx]
    ax.set_title(titulo, fontweight='bold', fontsize=11)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    
    # Desenhar refer√™ncia
    if ref_type == 's':
        # Eixo s (x=2, y=1)
        ax.axvline(x=2, color='red', linestyle='--', alpha=0.5, linewidth=2, label='Eixo s')
        ax.plot(2, 1, 'ro', markersize=10)
        ax.text(2.1, 1.1, 'Eixo s\n(2,1)', fontsize=9)
    elif ref_type == 'plane':
        # Visualizar alguns pontos do plano C
        q_vals = np.linspace(-0.5, 0.5, 15)
        p_vals = np.linspace(-0.5, 0.5, 15)
        plano_pts = []
        for q in q_vals:
            for p in p_vals:
                pt = P0_c + q*v1 + p*v2
                plano_pts.append(pt)
        plano_pts = np.array(plano_pts)
        ax.scatter(plano_pts[:, 0], plano_pts[:, 1], c='cyan', s=5, alpha=0.3, label='Plano C')
    elif ref_type == 'axis':
        # Eixo D
        t_vals = np.linspace(-2, 2, 50)
        eixo_pts = np.array([P0_d + t*v_unit for t in t_vals])
        ax.plot(eixo_pts[:, 0], eixo_pts[:, 1], 'r--', linewidth=2, alpha=0.5, label='Eixo D')
    
    # Cubo original (azul claro)
    for i, j in arestas:
        ax.plot([cubo[i][0], cubo[j][0]], 
                [cubo[i][1], cubo[j][1]], 
                'b-', alpha=0.4, linewidth=1.5)
    
    # V√©rtices originais
    ax.scatter(cubo[:, 0], cubo[:, 1], c='blue', s=30, alpha=0.5, zorder=5)
    
    # Cubo transformado (vermelho)
    cubo_trans = np.array([matriz @ np.append(v, 1) for v in cubo])[:, 0:3]
    
    for i, j in arestas:
        ax.plot([cubo_trans[i][0], cubo_trans[j][0]], 
                [cubo_trans[i][1], cubo_trans[j][1]], 
                'r-', linewidth=2.5)
    
    # V√©rtices transformados
    ax.scatter(cubo_trans[:, 0], cubo_trans[:, 1], c='red', s=40, zorder=5)
    
    # Setas de alguns v√©rtices
    for i in [6, 7]:  # V√©rtices superiores
        ax.annotate('', xy=(cubo_trans[i][0], cubo_trans[i][1]), 
                    xytext=(cubo[i][0], cubo[i][1]),
                    arrowprops=dict(arrowstyle='->', color='green', lw=1.5, alpha=0.7))
    
    # Legendas
    ax.plot([], [], 'b-', alpha=0.4, linewidth=1.5, label='Original')
    ax.plot([], [], 'r-', linewidth=2.5, label='Transformado')
    ax.legend(loc='best', fontsize=9)

plt.tight_layout()
plt.show()
```
